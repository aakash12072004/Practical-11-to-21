```python
import pandas as pd
import numpy as np

# STEP 1: Create Dirty Dataset
raw_data = {
 'Employee_ID': [101, 102, 103, 101, 105, 106],
 'Name': ['anil kumar', 'Sunita', np.nan, 'anil kumar', 'Rahul', 'amit'],
 'Age': [28, 32, 150, 28, np.nan, 45],
 'Salary': ['50000', '60000', '75000', '50000', 'missing', '45000'],
 'Join_Date': ['2023-01-10', '2023/05/12', '12-08-2023', '2023-01-10', '2023-09-01', '15/07/2023']
}

df = pd.DataFrame(raw_data)

print("--- Step 1: Raw Data ---")
print(df)

# STEP 2: Remove Duplicates
df = df.drop_duplicates()

print("\n--- Step 2: Duplicates Removed ---")
print(df)

# STEP 3: Handle Missing Values
df['Name'] = df['Name'].fillna('Unknown')

# Fill Age with mean
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

# STEP 4: Data Type Transformation
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

# Fill missing salary with median
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)

# STEP 5: Fix Outliers & Standardize Text
# Replace Age > 100 with average (30 for simplicity)
df.loc[df['Age'] > 100, 'Age'] = 30

# Capitalize names
df['Name'] = df['Name'].str.title()

# STEP 6: Normalize Date Format
df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors='coerce')

print("\n--- Step 6: Final Cleaned Data ---")
print(df)

# OPTIONAL: Save cleaned data
df.to_csv("cleaned_employee_data.csv", index=False)

print("\nCleaned dataset saved as 'cleaned_employee_data.csv'")
```
