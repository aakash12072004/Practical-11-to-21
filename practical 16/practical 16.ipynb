```python
import pandas as pd

# STEP 1: Semi-structured JSON data
data = [
 {"id": 1, "info": {"name": "Laptop", "price": 1200}, "tags": ["electronics", "office"]},
 {"id": 2, "info": {"name": "Chair", "price": 300}, "tags": ["furniture", "home"]},
 {"id": 3, "info": {"name": "Monitor", "price": 500}, "tags": ["electronics"]},
 {"id": 4, "info": {"name": "Desk", "price": 700}, "tags": ["furniture", "office"]},
 {"id": 5, "info": {"name": "Mouse", "price": 50}, "tags": ["electronics", "accessories"]}
]

# STEP 2: Manually Flatten JSON (NO json_normalize)
flat_data = []

for item in data:
    new_row = {
        "id": item["id"],
        "name": item["info"]["name"],
        "price": item["info"]["price"],
        "tags": item["tags"]
    }
    flat_data.append(new_row)

df_structured = pd.DataFrame(flat_data)

print("=== Step 3: Flattened Data ===")
print(df_structured)

# STEP 3: Manual explode (NO explode function)
rows = []
for _, row in df_structured.iterrows():
    for tag in row["tags"]:
        new_row = row.copy()
        new_row["tags"] = tag
        rows.append(new_row)

df_final = pd.DataFrame(rows)

print("\n=== Step 4: Fully Structured Data ===")
print(df_final)

# STEP 4: Summarization
print("\n=== Statistics (Price) ===")
print(df_final["price"].describe())

print("\n=== Average Price per Category ===")
print(df_final.groupby("tags")["price"].mean())

# STEP 5: Export CSV
df_final.to_csv("structured_output.csv", index=False)

print("\nFile 'structured_output.csv' created successfully!")
```
