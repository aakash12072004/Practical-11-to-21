```python
#seaborn and sns uske karine karvano baki che e karine lakhvo 

import pandas as pd
import matplotlib.pyplot as plt

# STEP 1: Dataset
data = {
 'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'] * 2,
 'Sales': [450, 480, 510, 550, 600, 650, 400, 420, 380, 410, 390, 400],
 'Product': ['Laptop'] * 6 + ['Tablet'] * 6,
 'Advertising_Spend': [10, 12, 15, 18, 20, 25, 8, 9, 7, 10, 8, 9]
}

df = pd.DataFrame(data)

print("--- Dataset ---")
print(df)

# STEP 2: LINE PLOT (Trend)
plt.figure()

for product in df['Product'].unique():
    subset = df[df['Product'] == product]
    plt.plot(subset['Month'], subset['Sales'], marker='o', label=product)

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()

# STEP 3: BAR CHART (Comparison)
plt.figure()

grouped = df.groupby('Product')['Sales'].sum()
plt.bar(grouped.index, grouped.values)

plt.title('Total Sales Comparison')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()

# STEP 4: SCATTER PLOT (Relationship)
plt.figure()

for product in df['Product'].unique():
    subset = df[df['Product'] == product]
    plt.scatter(subset['Advertising_Spend'], subset['Sales'], label=product)

plt.title('Advertising vs Sales')
plt.xlabel('Advertising Spend')
plt.ylabel('Sales')
plt.legend()
plt.show()

# STEP 5: BOX PLOT (Distribution)
plt.figure()

df.boxplot(column='Sales', by='Product')

plt.title('Sales Distribution')
plt.suptitle("")  # remove extra title
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```
